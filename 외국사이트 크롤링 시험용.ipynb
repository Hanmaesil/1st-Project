# 이미지 경로 추출하기

```python
import re #re임폴트
```

```python
pattern = re.compile('url\((.*?)\)') #url값만 추출하는 코드
#regex 구문 이해: 원하는 값을 둘러싸고 있는 값을 넣어주고 원하는 값은 
#(.*?)안에 넣은 후에 찿은 값을 group(1)으로 호출한 겁니다
```

```python
a ='<div class="image" style="background-image: url(https://ip-distiller.imgix.net/images/spirits/ecc017c0-8846-0135-bed5-0e6cfe74208a/zubrowka_bison_grass_vodka_thumbnail.jpg?w=120&amp;h=120&amp;mark-w=35&amp;mark-h=35&amp;fit=crop&amp;crop=right&amp;mark-align=top,left&amp;mark-pad=0&amp;mark=)"></div>'
#원하는 태그 선택
```

```python
found = re.search(pattern,a)
```

```python
print(found.group(1))
```
