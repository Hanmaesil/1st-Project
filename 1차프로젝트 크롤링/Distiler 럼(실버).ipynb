```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
import time
import random
import requests as req
from bs4 import BeautifulSoup as bs
import re
import pandas as pd
```

```python
pattern = re.compile('url\((.*?)\)') 
name_list = [] #이름
score_list = [] #별점
abv_list = [] #도수
des_list = [] #설명
tnote_list = [] #테이스팅 노트
taste_list = [] #맛, 향
img_list = [] # 이미지
num_list = [] #순서
num = 0

driver = wb.Chrome()
driver.maximize_window()
time.sleep(random.uniform(2,4))

for j in range(1,13):
    driver.get('https://distiller.com/search?page='+str(j)+'&sort=total_num_of_ratings&spirit_style_id=56')
    time.sleep(random.uniform(2,4))
    for i in range(10) :
        try:
            btn = driver.find_elements_by_class_name("name")
            btn[i].click()
            time.sleep(random.uniform(2,4))
            name = driver.find_element_by_css_selector("span.secondary-headline.name")
            name_list.append(name.text)
            try:
                score = driver.find_element_by_css_selector('#main-content > div > div.spirit-show.center-column.js-carousel.sugarcane-content > div.primary-content > div.tabs > div > div > div.stats-container > div > div > div > div > div.rating-display__value.average-rating > span')
                score_list.append(score.text)
            except:
                score_list.append("none")
            abv = driver.find_element_by_css_selector('#main-content > div > div.spirit-show.center-column.js-carousel.sugarcane-content > div.primary-content > div.tabs > div > div > div.spirit-show__description-container > div.other-details-container > ul > div > li.detail.abv > div.value')
            abv_list.append(abv.text)
            try:
                des = driver.find_element_by_css_selector('#main-content > div > div.spirit-show.center-column.js-carousel.sugarcane-content > div.primary-content > div.tabs > div > div > div.spirit-show__description-container > div.only-big > p')
                des_list.append(des.text)
                print("설명 입력")
            except:
                des_list.append("none")
                print("설명없음")
            try:
                tnote = driver.find_element_by_css_selector('#main-content > div > div.spirit-show.center-column.js-carousel.sugarcane-content > div.primary-content > div.tabs > div > div > div.details-module.spirit-show__tasting-notes.screened > div.content > p')
                tnote_list.append(tnote.text)
            except:
                tnote_list.append("none") 
            try :
                canvasTemp=driver.find_element_by_css_selector(".js-flavor-profile-chart")
                taste_list.append(canvasTemp.get_attribute('data-flavors'))
            except :
                taste_list.append("none")
                print("오류")
            url = driver.current_url
            res = req.get(url)
            found = re.search(pattern,res.text)
            img_list.append(found.group(1))
            num_list.append(num+1)
            num = num + 1
            driver.back()
            time.sleep(random.uniform(2,4))
        except:
            body = driver.find_element_by_css_selector("body")
            body.send_keys(Keys.ARROW_DOWN)
            body.send_keys(Keys.ARROW_DOWN)
            body.send_keys(Keys.ARROW_DOWN)
            body.send_keys(Keys.ARROW_DOWN)
            body.send_keys(Keys.ARROW_DOWN)
            body.send_keys(Keys.ARROW_DOWN)
            body.send_keys(Keys.ARROW_DOWN)
            body.send_keys(Keys.ARROW_DOWN)
            body.send_keys(Keys.ARROW_DOWN)
            body.send_keys(Keys.ARROW_DOWN)
            body.send_keys(Keys.ARROW_DOWN)
            body.send_keys(Keys.ARROW_DOWN)
            body.send_keys(Keys.ARROW_DOWN)
            body.send_keys(Keys.ARROW_DOWN)#
            body.send_keys(Keys.ARROW_DOWN)
            print("다운")
        
    
dic = {"순서" : num_list,"상품명" : name_list, "별점" : score_list, "도수" : abv_list, "설명" : des_list, "테이스트노트" : tnote_list, "맛과향" : taste_list, "이미지" : img_list}
fin = pd.DataFrame(dic)
fin
```

```python
fin.to_csv("distiller 럼 - 실버.csv", encoding="utf-8-sig")
```
